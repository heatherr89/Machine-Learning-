```python
#import dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#import cleaned data
data = pd.read_csv('clean_survey.csv')
data.head()
```

```python
type(data)
```

```python
# drop the timestamp
data.drop(columns=["Timestamp"],inplace = True)
data.head()
```

```python
#Encode the data for categorical variables
data = pd.get_dummies(data)
data.head()
```

```python
#features data 
features_data = data.drop(columns = ["mental_health_consequence_Maybe","mental_vs_physical_Don't know","mental_vs_physical_No","mental_vs_physical_Yes"])
features_labels = list(features_data.columns)
```

```python
#convert data into arrays

target_variable = np.array(data[["mental_health_consequence_Maybe","mental_vs_physical_Don't know","mental_vs_physical_No","mental_vs_physical_Yes"]])
features_variable = np.array(features_data)
```

```python
#review the shape
```

```python
#split the data into training and testing sets
train_features, test_features,train_target,test_target =train_test_split(features_variable,target_variable,test_size=0.25,random_state=42)
```

```python
#train the model
model = RandomForestRegressor(n_estimators = 1000, random_state = 42)
```

```python
model.fit(train_features,train_target)
```

```python
# Use the forest's predict method on the test data
predictions = model.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_target)
```

```python
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_target)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
# Get numerical feature importances
importances = list(model.feature_importances_)
```

```python
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(features_labels, importances)]
```

```python
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
```

```python
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

```python
# New random forest with only the two most important variables
rf_most_important = RandomForestRegressor(n_estimators= 1000, random_state=42)
```

```python
# Extract the two most important features
important_indices = [features_labels.index('mental_health_consequence_Yes'), features_labels.index('mental_health_consequence_No'),features_labels.index('Age')]
train_important = train_features[:, important_indices]
test_important = test_features[:, important_indices]
```

```python
# Train the random forest
rf_most_important.fit(train_important, train_target)
```

```python
# Make predictions and determine the error
predictions = rf_most_important.predict(test_important)
```

```python
errors = abs(predictions - test_target)
```

```python
# Display the performance metrics
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
mape = np.mean(100 * (errors / test_target))
accuracy = 100 - mape
print('Accuracy:', round(accuracy, 2), '%.')
```
